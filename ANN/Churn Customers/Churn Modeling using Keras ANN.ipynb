```python
import pandas as pd
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,3:13]
Y = df.iloc[:,13]
```

```python
X
```

```python
X = pd.get_dummies(X,drop_first= True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size= 0.6)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain_fit = sc.fit_transform(xtrain)
xtest_transform = sc.transform(xtest)
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,ReLU,Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(keras.Input(shape=(11,)))
```

```python
classifier.add(Dense(units = 6,activation = 'relu', kernel_initializer='he_uniform'))
```

```python
classifier
```

```python
classifier.add(Dense(units = 6,activation = 'relu', kernel_initializer='he_uniform'))
```

```python
classifier.add(Dense(units = 1,activation = 'sigmoid', kernel_initializer='glorot_uniform'))
```

```python
classifier.summary()
```

```python
#compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history = classifier.fit(x = xtrain_fit, y = ytrain,batch_size=10, epochs=100)
```

```python
ypred = classifier.predict(xtest_transform)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
ypred = (ypred > 0.5)
```

```python
cm = confusion_matrix(ytest,ypred)
```

```python
cm
```

```python
score = accuracy_score(ypred,ytest)
```
